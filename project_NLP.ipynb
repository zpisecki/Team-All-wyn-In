```python
tfidf = vectorizer.fit_transform(JobDesc['SOC Definition'])
```

```python
print(tfidf.toarray())
```

```python
from sklearn.metrics.pairwise import linear_kernel
```

```python
z = 0
dfResults = []
for z in range(0,11):
    tfidf[z:z+1]
    cosineSimilarities = linear_kernel(tfidf[z:z+1], tfidf).flatten()
    relatedJobIndices = cosineSimilarities.argsort()[:-12:-1]
    countMax = 1
    filterArr = []
    for j in relatedJobIndices:
        if (j > 10 and countMax < 4):
            filterArr.append(True)
            countMax = countMax + 1
        else:
            filterArr.append(False)
            
    newArr = relatedJobIndices[filterArr]
    jobIndices = np.array(newArr)
    cosScores = np.array(cosineSimilarities[newArr])
    d = pd.DataFrame({'jobIndices':jobIndices, 'cosScores':cosScores, 'occ code':JobDesc.iloc[jobIndices,2], 'occ title': JobDesc.iloc[jobIndices,0]}, columns=['jobIndices','cosScores', 'occ code', 'occ title'])
    d['Allwyn title'] = JobDesc.iloc[z,0]
    d['Allwyn Job Description'] = JobDesc.iloc[z,1]
    #d['occ code'] = JobDesc.iloc[d['jobIndices'],2]
    #d['occ title'] = JobDesc.iloc[d['jobIndices'],0]
    dfResults.append(d)
    print(dfResults)

print('concatenating data')
dt = pd.concat(dfResults, ignore_index=True) 
del dfResults
```
